```python
import pandas as pd
import numpy as np
from scipy.stats import shapiro
from sklearn.preprocessing import MinMaxScaler
import torch
from torch import nn
from torch.autograd import Variable
```

```python
features = pd.read_csv('train_data/train_expression.csv')
labels = pd.read_csv('train_data/train_labels.csv')
```

```python
size_features = len(features.columns)
```

```python
corr = [labels['age'].corr(features[features.columns[i]]) for i in range(size_features)]
sorted_corr = [i for i in sorted(enumerate(corr), key=lambda x:abs(x[1]), reverse=True)]
```

```python
features_strong_index = [i[0] for i in sorted_corr if abs(i[1]) > 0.05]
```

```python
features_strong = features[[features.columns[i] for i in features_strong_index]]
```

```python
size_features_strong = len(features_strong.columns)
p = [shapiro(features_strong[features_strong.columns[i]])[1] for i in range(size_features_strong)]
```

```python
scaler = MinMaxScaler()
features_strong_scaled = scaler.fit_transform(features_strong)
```

```python
class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork, self).__init__()
        self.linear_sigmoid_stack = nn.Sequential(
            nn.Linear(size_features_strong, 5000),
            nn.Sigmoid(),
            nn.Linear(5000, 500),
            nn.Sigmoid(),
            nn.Linear(500, 50),
            nn.Sigmoid(),
            nn.Linear(50,1)
        )

    def forward(self, x):
        out = self.linear_sigmoid_stack(x)
        return out
```

```python
model = NeuralNetwork()
criterion = torch.nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=1e-4)
```
